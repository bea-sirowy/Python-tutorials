## Imports: storing and retrieving modules

A module is a file with the .py extension containing Python code.

```python
# the following code will be saved as my_module.py
PI = 3.14159

def f(x):
    return x * 2

def g(a, b, c):
    return a + b + c
```

We can access the variables and functions defined in my_module.py, from another file in the same directory:

```python
import my_module as mm
result = mm.f(5)
pi = mm.PI
```

Alternatively:


```python
from my_module import g as gf, PI as pi
result = gf(5, pi, 7)
```

The backslash character \ is an escape character - it is used to specify special characters like newline \n or Unicode characters. To write a string literal with backslashes, we need to escape them:

```python
s = "12\30"
print(s)
```

```python
s = "12\\30"
print(s)
```

Alternatively, you can preface the leading quote of the string with r (__r stands for raw__), which means that the characters should be interpreted as is:

```python
 s = r"12\30"
 d = r"this\has\no\special\characters"
 print(s)
 print(d)
```

f-strings: formatted string literals. To create an f-string, write the character f immediately preceding a string literal. Within the string, enclose Python expressions in curly braces to substitute the value of the expression into the formatted string:

```python
amount = 10

rate = 88.46

currency = "Pesos"

result = f"{amount} {currency} is worth US${amount / rate}"

print(result)
```

adding __:.2f__ formats the first argument as a floating-point number with two decimal places

```python
amount = 10

rate = 88.46

currency = "Pesos"

result = f"{amount} {currency} is worth US${amount / rate:.2f}"

print(result)
```

### None: null value type

None is a common default value for function arguments:

```python
def add_and_maybe_multiply(a, b, c=None):
    result = a + b

    if c is not None:
        result = result * c

    return result

print(add_and_maybe_multiply(3, 4))
print(add_and_maybe_multiply(3, 4, 7))
```

```python
sequence = [1, 2, None, 4, None,]
total = 0
for n in sequence:
    print (n)
    if n is None:
        continue
    total += n 
print ("Total:", total)
```

```python
sequence = [1, 2, 0, 8, 25, 2, 5, 1]
total_until_5 = 0
for n in sequence:
    print (n)
    if n is 5:
        break
    total_until_5 += n 
print ("Total_until_5:", total_until_5)
```
