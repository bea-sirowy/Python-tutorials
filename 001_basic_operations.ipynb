## Imports: storing and retrieving modules

A module is a file with the .py extension containing Python code.

```python
# the following code will be saved as my_module.py
PI = 3.14159

def f(x):
    return x * 2

def g(a, b, c):
    return a + b + c
```

We can access the variables and functions defined in my_module.py, from another file in the same directory:

```python
import my_module as mm
result = mm.f(5)
pi = mm.PI
```

Alternatively:


```python
from my_module import g as gf, PI as pi
result = gf(5, pi, 7)
```
