## An overview
- a programming language of "marker characters"
- "old school" - very compact
- a way to reduce the amont of code - instead of programming with lines we program with characters
- used also with other languages - e.g. SQL

## RE in Python

- we must import the library using "import re"
- we can use __re.search()__  to see if a stringf matches a regular expression, similar to using the __find()__ or __startswith()__ methods for strings - see the examples below

```python
document=open(r"C:\Users\Beata\Documents\Python Scripts\mbox-short.txt")
for line in document:
    line = line.rstrip()
    if line.find('From: g') >= 0:
        print(line)
```

```python
document=open(r"C:\Users\Beata\Documents\Python Scripts\mbox-short.txt")
for line in document:
    line = line.rstrip()
    if line.startswith('From: g'):
        print(line)
```

```python
import re

document=open(r"C:\Users\Beata\Documents\Python Scripts\mbox-short.txt")
for line in document:
    line = line.rstrip()
    if re.search('From: g', line):
        print(line) 
```

```python
import re

document=open(r"C:\Users\Beata\Documents\Python Scripts\mbox-short.txt")
for line in document:
    line = line.rstrip()
    if re.search('^From: g', line): #Here we use a special character ^ (starts with) 
        print(line) 
```

- we can use __re.findall()__ to extract portions of a string that match our regular expression, similar to a combination of find() and slicing: var[5:10]
